```python
import pandas as pd
```

```python
data1 = pd.read_csv('data/owid-covid-data.csv')
justAprilData = data1['date']=='2022-03-01'
covidData_April2022 = data1[justAprilData]
```

```python
covidData_April2022.head(5)
```

```python
covidData_April2022 = covidData_April2022.set_index("date")
```

```python
list(covidData_April2022.columns)
```

```python
filteredCovidData = [covidData_April2022['iso_code'],
                      covidData_April2022['continent'],
                      covidData_April2022['location'],
                      covidData_April2022['total_cases'],
                      covidData_April2022['total_deaths'],
                      covidData_April2022['total_vaccinations'],
                      covidData_April2022['people_vaccinated'],
                      covidData_April2022['total_vaccinations_per_hundred'],
                      covidData_April2022['people_vaccinated_per_hundred'],
                      covidData_April2022['population'],
                      covidData_April2022['population_density'],
                      covidData_April2022['median_age']]

headers = ["iso_code",
           "continent",
           "location",
           "total_cases",
           "total_deaths",
           "total_vaccinations",
           "people_vaccinated",
           "total_vaccinations_per_hundred",
           "people_vaccinated_per_hundred",
           "population",
           "population_density",
           "median_age"]


newCovidData= pd.concat(filteredCovidData, axis=1, keys=headers)
```

```python
newCovidData.shape
```

```python
import matplotlib.pyplot as plt
```

```python
newCovidData = newCovidData.set_index("people_vaccinated_per_hundred")
```

```python
plt.scatter(newCovidData.index, newCovidData["people_vaccinated_per_hundred"])
```
