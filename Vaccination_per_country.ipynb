```python
data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d')
```

```python
filtered_df = data.loc[(data['date'] >= '2021-01-01')
                     & (data['date'] < '2022-04-01')]
```

```python
filtered_df.dtypes
```

```python
list(filtered_df.columns)
```

```python
filteredCovidData = [filtered_df['iso_code'],
                     filtered_df['date'],
                      filtered_df['continent'],
                      filtered_df['location'],
                      filtered_df['total_cases'],
                      filtered_df['total_deaths'],
                      filtered_df['total_vaccinations'],
                      filtered_df['people_vaccinated'],
                      filtered_df['total_vaccinations_per_hundred'],
                      filtered_df['people_vaccinated_per_hundred'],
                      filtered_df['population'],
                      filtered_df['population_density'],
                      filtered_df['median_age']]

headers = ["iso_code",
           "date",
           "continent",
           "location",
           "total_cases",
           "total_deaths",
           "total_vaccinations",
           "people_vaccinated",
           "total_vaccinations_per_hundred",
           "people_vaccinated_per_hundred",
           "population",
           "population_density",
           "median_age"]


newCovidData= pd.concat(filteredCovidData, axis=1, keys=headers)
```
